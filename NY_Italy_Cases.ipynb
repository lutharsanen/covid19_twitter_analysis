```python
import os
import pandas as pd
import matplotlib.pyplot as plt

#USA data from https://github.com/nytimes/covid-19-data 
df = pd.read_csv('us-counties.csv', sep=',')

#italy data from https://github.com/pcm-dpc/COVID-19/tree/master/dati-regioni 
df_ita = pd.read_csv('dpc-covid19-ita-regioni.csv', sep=',')
```

```python
os.getcwd()
```

```python

df.groupby('state').count()
```

```python
df_ny = df[df['state'].str.contains("New York")]
df_ny.groupby('county').sum()
```

```python
df_nyc = df[df['county'].str.contains("New York City")]
df_nyc.head()
```

```python
#NY versus NYC (out of interest)
plt.plot(df_ny.date, df_ny.cases, color='g')
plt.plot(df_nyc.date, df_nyc.deaths, color='b')
plt.plot(df_ny.date, df_ny.deaths, color='r')


plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('New York State Covid-19 Cases 01.03 - 17.04')
plt.xscale
plt.show()
```

```python
## italy starting 20-02-24
df_ita.head(15)
df_ita.groupby('denominazione_regione').count()
```

```python
plt.plot(df_ita.data, df_ita.totale_casi, color='g')
plt.plot(df_ita.data, df_ita.deceduti, color='r')

plt.xlabel('Date')
plt.ylabel('Cases')
plt.title('Italy Covid-19 Cases 24.02 - 17.04')
plt.show()
```
